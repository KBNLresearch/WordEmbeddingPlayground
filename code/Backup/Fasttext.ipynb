## Training and saving a model

```python
sentences = SentIterator("./data/kranten_pd_1875-6.zip")
```

```python
from gensim.models.fasttext import FastText 
model = FastText(size=100, window=5, min_count=10)
model.build_vocab(sentences=sentences)
total_examples = model.corpus_count
print(total_examples)
model.train(sentences=sentences, total_examples=total_examples, epochs=5)
```

```python
path = "./models/kranten_pd_1875-6_model.fasttext"
model.save(path)
```

## Loading a model

```python
model = FastText.load(path)
print(model)
```

```python
query = 'vrouw'
print(query in model.wv.vocab)
model.most_similar(query,topn=20)
```
