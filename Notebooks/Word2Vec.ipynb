## Training and saving a model

```python
sentences = SentIterator("./data/kranten_pd_1875-6.zip")
```

```python
from gensim.models.word2vec import Word2Vec 
model = Word2Vec(size=100, window=5, min_count=10)
model.build_vocab(sentences=sentences)
total_examples = model.corpus_count
print(total_examples)
model.train(sentences=sentences, total_examples=total_examples, epochs=5)
```

```python
path = "./models/kranten_pd_1875-6_model.word2vec"
model.save(path)
```

## Loading a model

```python
model = Word2Vec.load(path)
print(model)
```

```python
query = 'vrouw'
model.most_similar(query)
```
